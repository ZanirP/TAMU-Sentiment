```python
# System Imports
import sys
import os
sys.path.append(os.path.join(os.path.abspath('../src')))

# Personal function imports
from preprocessing_data.events import common_event_terms
from preprocessing_data.events import preprocess_text


# Machine Learning Imports
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity



```

STEP 2: Preprocess the text! \
STEP 3: Vectorize with TF-IDF \
STEP 4: Compute Similarity \

```python
common_event_terms = ' '.join(common_event_terms)

common_event_terms = preprocess_text(common_event_terms)
common_event_terms = ' '.join([word for word in common_event_terms.split() if word not in ENGLISH_STOP_WORDS]) # Remove stop words
# shouldn't be necessary, but just in case


documents = [common_event_terms, "florida gators vs texas aampm aggies prediction odds say coach billy napi via"]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

print(tfidf_matrix.toarray())
```

```python
similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
print(f"Cosine similarity between event text and new text: {similarity[0][0]}")
```
